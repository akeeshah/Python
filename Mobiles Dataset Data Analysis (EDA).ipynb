### Data cleaning and formatting

```python
df['Company Name'] = df['Company Name'].str.lower().str.title()

unique_CompanyName = df['Company Name'].unique()
print(unique_CompanyName)
```

```python
df.info()
```

```python
# Drop Duplicates
df.drop_duplicates(inplace=True)
```

```python
# Strip spaces from column names
df.columns = df.columns.str.strip()
```

```python
df.head()
```

```python
# list od price columns

price_columns = [
    'Launched Price (Pakistan)',
    'Launched Price (India)',
    'Launched Price (China)',
    'Launched Price (USA)',
    'Launched Price (Dubai)'
]

# Clean each price Column

for col in price_columns:
    df[col] = (
        df[col]
        .replace(r'[^\d.]', '', regex=True)
        .replace('',np.nan)
        .astype(float)
    )
```

```python
df.head()
```

### Missing Value Analysis

```python
missing = df.isnull().sum()
missing_percent = (missing / len(df)) * 100
missing_df = pd.DataFrame({'Missing Values':missing, 'Percentage':missing_percent})
missing_df[missing_df['Missing Values']>0]
```

### Univariate Analysis

```python
sns.histplot(df['Launched Price (India)'],bins=50,kde=True)
plt.title("Price Distribution")
plt.show()
```

```python
for col in price_columns:
    sns.histplot(df[col],bins=50,kde=True)
    plt.title(f"Price Distribution: [col]")
    plt.xlabel('Price')
    plt.ylabel('Frequency')
    plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(y='Company Name', data=df, order=df['Company Name'].value_counts().index)
plt.title("Mobile Brands Count")
plt.show()
```

```python
def extract_first_num(s):
    import re
    match = re.search(r'[\d]+',str(s))
    return float(match.group()) if match else None

df['RAM (GB)'] = df['RAM'].apply(extract_first_num)
df['Front Camera (MP)'] = df['Front Camera'].apply(extract_first_num)
df['Back Camera (MP)'] = df['Back Camera'].apply(extract_first_num)
df['Mobile Weight (g)'] = df['Mobile Weight'].apply(extract_first_num)
df['Battery Capacity (mAh)'] = df['Battery Capacity'].str.replace(',','').apply(extract_first_num)
df['Screen Size (inches)'] = df['Screen Size'].str.extract(r'(\d+\.?\d*)').astype(float)
```

### Weight vs Battery vs Screen Size (3D Analysis)

```python
fig = plt.figure(figsize=(10,7))
ax=fig.add_subplot(111,projection='3d')
ax.scatter(
    df['Mobile Weight (g)'],
    df['Battery Capacity (mAh)'],
    df['Screen Size (inches)'],
    c='green', alpha=0.6
)
ax.set_xlabel('Weight (g)')
ax.set_ylabel('Battery (mAh)')
ax.set_zlabel('Screen Size (inches)')
plt.title('3D Analysis: Weight vs Battery vs Screen Size')
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

fig = px.scatter_3d(
    df,
    x='Mobile Weight (g)',
    y='Battery Capacity (mAh)',
    z='Screen Size (inches)',
    color='Battery Capacity (mAh)',
    size='Screen Size (inches)',
    opacity=0.7,
    title='3D Analysis: Weight vs Battery vs Screen Size',
    labels={
        'Mobile Weight (g)': 'Weight (g)',
        'Battery Capacity (mAh)': 'Battery (mAh)',
        'Screen Size (inches)': 'Screen Size (inches)'
    }
)

fig.update_traces(marker=dict(line=dict(width=0.5,color='DarkSlateGrey')))
fig.update_layout(margin=dict(l=0, r=0, b=0, t=40))
fig.show()
```
